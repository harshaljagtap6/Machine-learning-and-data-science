```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset.head()
```

```python
x = dataset.iloc[:, [3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
ddg = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_ac = ac.fit_predict(x)
print(y_ac)
```

```python
plt.scatter(x[y_ac == 0, 0], x[y_ac == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(x[y_ac == 1, 0], x[y_ac == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_ac == 2, 0], x[y_ac == 2, 1], c = 'green', label = 'Cluster 3')
plt.scatter(x[y_ac == 3, 0], x[y_ac == 3, 1], c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_ac == 4, 0], x[y_ac == 4, 1], c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
